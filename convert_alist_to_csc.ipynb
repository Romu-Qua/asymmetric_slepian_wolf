# Simple conversion of alist format to sparse column format


This notebook executes a simple converison of a parity check matrix given in the alist format
(see https://www.inference.org.uk/mackay/codes/data.html) into a sparse column format,
outputing the respective column pointers and row indeces into numpy file for further processing

```python
import numpy as np
import scipy.sparse as spr
import sys
import argparse
from tqdm import tqdm
```

Replace the path with your specific saved file

```python
#filepath = "/home/bob/Documents/information_theory/4095_737_3_101.txt"
#filepath = "/home/bob/Documents/information_theory/4095_738_4_102.txt"
filepath = "/home/bob/Documents/information_theory/1908_212_4_1383.txt"
```

Assing the column to row switch for the chosen matrix. This is the line in
which the column description switches to the row descritpion in the alist file
(this means where the rows get longer/shorter)

```python
column_to_row_switch = 1913

with open(filepath) as f:
    lines = f.readlines()
    for num, line in enumerate(lines):
        #print(line)
        line = line.strip()
        numbers = line.split(" ")
        #print(numbers)
        #print("num", num)
        numbers = [int(a) for a in numbers]
        numbers = np.array(numbers, dtype=int)
        if num == 0:
            print("assigning n")
            n = int(numbers)
        elif num == 1:
            m = int(numbers)
            H = spr.lil_matrix((m, n), dtype=np.uint8)
        elif num == 2:
            pass
        elif num>=column_to_row_switch:
            print("num ",num)
            print(numbers)
            for value in numbers:
                if value == 0:
                    continue
                row = num - column_to_row_switch
                print(f"assingin row {row} with {numbers - 1}")
                H[row, value-1] = 1
                # H[num,numbers-1]=1
```

# saving the files to numpy arrays

```python
H_csc = H.tocsc()
row_index = H_csc.indices
column_pointers = H_csc.indptr

np.save("1908_212_4_colmn_pointers.npy", column_pointers.astype('uint32'))
np.save("1908_212_4_row_index.npy", row_index.astype('uint16'))
```
