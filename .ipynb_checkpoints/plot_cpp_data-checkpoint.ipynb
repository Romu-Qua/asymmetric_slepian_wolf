```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


#with open(path_fer, 'r') as f:
#    fers_str = f.read().splitlines()
    
#with open(path_ps, 'r') as f:
#    p_str = f.read().splitlines()

def plot_data(path_fer, path_ps):
    with open(path_fer, 'r') as f:
        fers_str = f.read().splitlines()
    
    with open(path_ps, 'r') as f:
        p_str = f.read().splitlines()
    
    fers = [float(num) for num in fers_str]
    ps = [float(num) for num in p_str]
    
    fers = np.array(fers)
    ps = np.array(ps)
    
    nonzeros = np.nonzero(fers>0)[0]
    fers = fers[nonzeros]
    ps = ps[nonzeros]
    
    print(ps)
    
    plt.scatter(ps, fers)
    plt.plot(ps[:12],fers[:12])
    plt.xscale('log')
    plt.yscale('log')
    plt.grid(which='both')
    plt.xlabel("Crossover Probability")
    plt.ylabel("FER")
    
#fers_str = fers_str+fers_detail_str
#p_str = p_str+p_detail_str
```

# Plot for 4095.737.101

```python
path_fer = "fer"
path_ps = "p"

plot_data(path_fer, path_ps)
```

# plot for 4095.738.102

```python
path_fer = "fer_detail_4095_738_102"
path_ps = "p_detail_4095_738_102"

plot_data(path_fer, path_ps)
```
